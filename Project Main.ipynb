```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
# Study data files
billboard_data = "Cleaned_Billboard.csv"
spotify_data = "data/adj_spotify_data.csv"
lyrics_data = "lyrics_final.csv"

# Read the mouse data and the study results
billboard_data = pd.read_csv(billboard_data)
spotify_data = pd.read_csv(spotify_data)
lyrics_data = pd.read_csv(lyrics_data)

lyrics_data.head()

spotify_data.rename(columns = {'name':'Name'}, inplace=True)
#spotify_data.head()
```

```python
lyrics_data.head()
```

```python
billboard_data.drop_duplicates('Name', inplace=True)
```

```python
merged_data_df = pd.merge(billboard_data, spotify_data, how='right', on='Name')
```

```python
new_merged_data_df = pd.merge(merged_data_df, lyrics_data, how='right', on='Name')
new_merged_data_df.head(20)
```

```python
new_merged_data_df.drop(columns=['Weekly.rank'], inplace=True)
new_merged_data_df.drop(columns=['Weeks.on.chart'], inplace=True)
new_merged_data_df.drop(columns=['Unnamed: 0'], inplace=True)
new_merged_data_df.drop(columns=['Genre'], inplace=True)
```

```python
new_merged_data_df.drop(columns=['New_Artist'], inplace=True)
new_merged_data_df.sort_values(by=['year'], ascending=False)
```

```python
new_merged_data_df.rename(columns = {'artists':'Artists', 'danceability':'Danceability', 'energy':'Energy', 'popularity':'Popularity', 'year':'Year'}, inplace=True)
new_merged_data_df.head()
```

```python
# Create database connection
rds_connection_string =(f'postgres:{password}@localhost:5432/Song_db')
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
new_merged_data_df.to_sql(name='premise', con=engine, if_exists='append', index=False)
```
