```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

### Store CSV into DataFrame

```python
csv_file = "../Resources/billboard.csv"
billboard_data_df = pd.read_csv(csv_file)
billboard_data_df.head()
```

### Create new data with select columns

```python
new_billboard_data_df = billboard_data_df[['Artists', 'Name', 'Weekly.rank', 'Weeks.on.chart', 'Date', 'Genre']].copy()
new_billboard_data_df.head()
```

### Clean DataFrame

```python
new_billboard_data_df["New_Genre"] = new_billboard_data_df["Genre"].str.split(",",expand=True)[0]
new_billboard_data_df["New_Artist"] = new_billboard_data_df["Artists"].str.split(",",expand=True)[0]
new_billboard_data_df.head()
```

```python
new_billboard_data_df = new_billboard_data_df.drop(columns=["Artists", "Genre"])
new_billboard_data_df.head()
```
